```python
import pandas as pd
import numpy as np
```

```python
fitbit = pd.read_csv('steps_df.csv')
apple = pd.read_csv('total_steps.csv')
```

```python
fitbit.rename({'value':'Fitbit_Steps'},axis=1, inplace=True)
```

```python
apple.rename({'value':'Apple_Steps'},axis=1, inplace=True)
```

```python
fitbit[fitbit.duplicated()==True]
```

```python
apple[apple.duplicated()==True]
```

```python
combined = fitbit.merge(apple,how='left',on='dateTime')
```

```python
combined['steps']=np.where(combined['Fitbit_Steps'] == 0, combined['Apple_Steps'], combined['Fitbit_Steps'])
```

```python
combined
```

```python
import plotly.express as px
```

```python
fig = px.line(combined, x='dateTime', y='steps', title='Time Series Plot')
# Add range slider
fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label="1m",
                     step="month",
                     stepmode="backward"),
                dict(count=6,
                     label="6m",
                     step="month",
                     stepmode="backward"),
                dict(count=1,
                     label="YTD",
                     step="year",
                     stepmode="todate"),
                dict(count=1,
                     label="1y",
                     step="year",
                     stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)

fig.show()
```
